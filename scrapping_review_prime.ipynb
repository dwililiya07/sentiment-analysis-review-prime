```python
# import library google_play_scraper untuk akses ulasan di google play store
from google_play_scraper import Sort, reviews
import pandas as pd
import csv
```

Selanjutnya dilakukan pengambilan data review yang paling baru untuk menghindari bias (karena sudah updatenya aplikasi) dari ulasan lama yang sudah tidak relevan. Kemudian data yang disimpan yaitu content, score, at, dan thumbsupcount.

```python
# melakukan scraping data sebanyak 15000 review dari aplikasi Prime Video di Google Play Store
app_id = 'com.amazon.avod.thirdpartyclient'
result, _ = reviews(
    app_id,
    lang='id',
    country='id',
    sort=Sort.NEWEST,
    count=15000
)

# simpan datasetnya
with open('reviews_primevideo.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file, quoting=csv.QUOTE_ALL)
    writer.writerow(['content', 'score', 'at', 'thumbsUpCount'])
    for r in result:
      writer.writerow([r['content'], r['score'], r['at'], r['thumbsUpCount']])
```

quoting=csv.QUOTE_ALL digunakan untuk memastikan file csv aman dari karakter spesial atau koma di dalam teks ulasan.
